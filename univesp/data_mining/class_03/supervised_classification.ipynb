```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
comedians = pd.read_csv('./shows.csv')
comedians.head(15)
```

```python
comedians.Nationality = comedians.Nationality.map({'UK': 0, 'USA': 1, 'N': 2})
comedians.Go = comedians.Go.map({'NO': 0, 'YES': 1})

comedians.head(15)
```

```python
attrs_names = ['Age', 'Experience', 'Rank', 'Nationality']
attrs = comedians[attrs_names]
classes = comedians.Go
```

```python
arvore = DecisionTreeClassifier()
arvore = arvore.fit(attrs.values, classes.values)
```

```python
plt.figure(figsize=(15, 8))
plot_tree(arvore, filled=True, rounded=True, class_names=['No', 'Yes'], feature_names=attrs_names)
plt.show()
```

```python
arvore.predict([[25, 10, 7, 0]])
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
wines = pd.read_csv(url, sep=';')
wines.head(20)
```

```python
wines.describe()
```

```python
wines.quality.unique()
```

```python
wines_columns = wines.columns[0:11].values.tolist()
wines_attrs = wines[wines_columns]
wines_classes = wines.quality
```

```python
attrs_training, attrs_test, wines_classes_training, wines_classes_test = train_test_split(wines_attrs, wines_classes, test_size=.1)
```

```python
wine_tree = DecisionTreeClassifier()
wine_tree = wine_tree.fit(attrs_training, wines_classes_training)
```

```python
wines_classes_prediction = wine_tree.predict(attrs_test)
accuracy = accuracy_score(wines_classes_test, wines_classes_prediction)
```

```python
print(f'Classification Accuracy: {accuracy}')
```

```python
plt.figure(figsize=(20, 12))
plot_tree(wine_tree, filled=True, rounded=True, class_names=['3', '4', '5', '6', '7', '8'])
```
