```python
precos = pd.DataFrame([['Feira', 'Cebola', 2.5], 
                        ['Mercado', 'Cebola', 1.99], 
                        ['Supermercado', 'Cebola', 1.69], 
                        ['Feira', 'Tomate', 4], 
                        ['Mercado', 'Tomate', 3.29], 
                        ['Supermercado', 'Tomate', 2.99], 
                        ['Feira', 'Batata', 4.2], 
                        ['Mercado', 'Batata', 3.99], 
                        ['Supermercado', 'Batata', 3.69]], 
                        columns = ['Local', 'Produto', 'Preço'])
precos
```

```python
produtos = precos.groupby('Produto', sort = False)
produtos.describe().round(2)
```

```python
produtos.Preço.aggregate(['mean', 'std', 'min', 'max'])\
    .rename(columns={'mean': 'Média', 'std': 'Desvio', 'min': 'Valor Minimo', 'max': 'Max'})\
    .round(3)
```
